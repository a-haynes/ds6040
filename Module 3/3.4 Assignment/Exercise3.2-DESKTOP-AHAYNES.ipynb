2. The code for the LDA and QDA classes are the same except for two
lines (one line in the init method and one in the compute likelihoods
method). What are these two lines, and what is the functional/mathematical
differences in what they do.

For the init method, we have the LDA code:

```
            self.cov = self.cov + self.num_obs[name] * np.cov(np.transpose(self.data[self.data_labels == name, :]))
        self.cov = self.cov / self.num_rows
```
and the QDA code:

```
            self.covs[name] = np.cov(np.transpose(self.data[self.data_labels == name, :]))
```

This is because for LDA we take the mean covariance matrix and share it across all classes, but QDA uses a covariance matrix per class.

For the likehood method, the LDA code:

```
            likelihoods[idx] = multivariate_gaussian_pdf(x, self.means[name], self.cov)
```

And the QDA code:

```
            likelihoods[idx] = multivariate_gaussian_pdf(x, self.means[name], self.covs[name])
```
While determine the likelihoods, the LDA method uses that shared covariance matrix we came up with in the last step while the QDA has each class use the class-specific covariance.
